```python
import pandas as pd
import numpy as np
```

```python
link = 'https://www.czso.cz/documents/10180/91917302/130055190105a.xlsx/ba744512-9c05-47ad-a6b0-3f5e9c0254d2?version=1.1'
data = pd.read_excel(link, header = 0, skiprows = 1, index_col = 0)
```

```python
data.iloc[1]
```

```python
col1 = data.iloc[0].fillna(method = 'ffill').str.split(pat='\n', n = 0).str.get(1)
#vyplni Nan predchozim objektem v listu
#rozdeli retezec podle \n a vezme druhou cast

col2 = data.iloc[1].str.split(pat='\n', n = 0).str.get(1)
columns = [col1, col2]

columns = list(zip(*columns))
# spoji oba listy do listu listu

columns
```

```python
data.columns = columns
data = data.drop(data.index[[0,1]])
data
```

```python
data.columns = pd.MultiIndex.from_tuples(data.columns, names=['Stat','Subdivision'])
data
```
