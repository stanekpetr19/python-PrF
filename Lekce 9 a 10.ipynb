```python
# Seskupování dat - groupby
#==========================

# Vytvoření podmnožiny dat 
# Groupby vytvoří iterátor, pomocí kterého můžeme procházet prvky 
# (v tomto případě části DataFrame rozděleného dle hodnoty sloupce 'sex')
# potom na tyto prvky aplikujeme metody mean()
titanic.groupby('sex').mean()
```

```python
titanic.groupby('pclass').mean()
titanic.groupby(['sex', 'pclass']).mean()
```

```python
titanic.groupby("pclass").survived.mean()
```
